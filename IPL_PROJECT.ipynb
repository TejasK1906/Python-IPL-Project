**Cleaning null values from dataframe**

```python
matches_ipl.drop(["method"],axis=1,inplace=True)
```

```python
ball_ipl["extra_type"].fillna("No Extra",inplace=True)
ball_ipl["player_out"].fillna('NotOut',inplace=True)
ball_ipl["kind"].fillna(0,inplace=True)
ball_ipl["fielders_involved"].fillna(0,inplace=True)
```

## Stats 

### Highest Individual Score

```python
H= ball_ipl.groupby(['batter','BattingTeam'])[("batsman_run")].sum().reset_index()
Pj = H.sort_values(by ="batsman_run",ascending= False).head(10).reset_index().drop('index', axis=1)
Pj.index = Pj.index + 1
Pj
```

```python
plt.figure(figsize=(12, 6))
plt.bar(Pj['batter'],Pj['batsman_run'])
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Total Runs in a Season');
```

### Highest Score In a Match

```python
C = ball_ipl.groupby(["batter",'ID',"BattingTeam"])[('batsman_run')].sum().reset_index()
Pn = C.groupby(["batter","BattingTeam","ID"])[("batsman_run")].max().reset_index().sort_values(by = "batsman_run",ascending = False).reset_index().head(10).drop('index',axis = 1).rename(columns={'batsman_run': "Highest Score"}).drop("ID",axis=1)
Pn.index = Pn.index + 1
Pn
```

### Highest Average In Tournament

```python
A = ball_ipl.groupby(['batter','BattingTeam'])[('isWicketDelivery','batsman_run')].sum().reset_index().sort_values(by = 'batsman_run',ascending = False).head(100).reset_index()
A["Average"] = A['batsman_run']/A['isWicketDelivery']
Bx = A.sort_values(by = 'Average', ascending = False).reset_index().head(10).drop('index', axis= 1)
Bx.index = Bx.index +1
Bx.drop('level_0', axis = 1)
```

### Highest Team Scores

```python
K = ball_ipl.groupby(['ID','BattingTeam','innings'])[('total_run'),("isWicketDelivery")].sum().reset_index()
K['Final_Score'] = K['total_run'].astype(str) + '/'+ K['isWicketDelivery'].astype(str)
J = K.sort_values(by = 'total_run',ascending= False).drop(['ID','total_run','isWicketDelivery'],axis = 1).reset_index()
P = J.drop('index',axis=1).head(15)
P.index = P.index +1
P
```

### Match Result

```python
A = ball_ipl.groupby(['ID',"BattingTeam",'innings'])[('total_run')].sum().reset_index()
B = A.loc[A['innings']==1].rename(columns = {'total_run':'Team1_score','BattingTeam':'Team1'})
C = ball_ipl.groupby(['ID',"BattingTeam",'innings'])[('total_run')].sum().reset_index()
D = A.loc[A['innings']==2].rename(columns= {'total_run':"Team2_score",'BattingTeam':'Team2'})
stand = B[['ID','Team1','innings','Team1_score']].merge(D,left_on = 'ID',right_on='ID',how='left').drop('ID',axis=1)
stand
stand.loc[stand['Team1_score'] > stand['Team2_score'] , 'Winner'] = stand['Team1']  
stand.loc[stand['Team1_score'] < stand['Team2_score'] , 'Winner'] = stand['Team2']
stand.index= stand.index +1
stand
```

### Highest Match Aggregate

```python
agg = stand[['Team1','Team2']]
agg['Total_match_score'] = stand['Team1_score']+stand['Team2_score']
z = agg.sort_values(by = 'Total_match_score',ascending = False).reset_index().drop('index',axis=1)
z.index = z.index + 1
z
```

### Largest Victory By Runs

```python
a = stand[['Team1',"Team2",'Winner']]
a['Margin'] = stand['Team1_score']-stand['Team2_score']
a['Victory Margin']= a['Margin'].abs()
large = a.sort_values(by = 'Victory Margin',ascending = False).drop('Margin',axis=1).head(15).reset_index().drop('index',axis=1)
large.index= large.index +1
large
```

### Most Sixes in the Tournament

```python
ball_ipl.loc[ball_ipl['batsman_run']==6,'Is Six'] = 1
six = ball_ipl.groupby(['batter','BattingTeam'])[('Is Six')].sum().reset_index().sort_values(by = 'Is Six',ascending= False).head(10).reset_index().drop('index',axis=1)
six['Is Six']= six['Is Six'].astype(int)
six.index = six.index  +1
six
```

### Most No of Extras Given by a Team

```python
A = ball_ipl.groupby(['ID',"BattingTeam",'innings'])[('total_run',"extras_run")].sum().reset_index()
B = A.loc[A['innings']==1].rename(columns = {'total_run':'Team1_score','BattingTeam':'Team1'})
C = ball_ipl.groupby(['ID',"BattingTeam",'innings'])[('total_run',"extras_run")].sum().reset_index()
D = A.loc[A['innings']==2].rename(columns= {'total_run':"Team2_score",'BattingTeam':'Team2'})
stand1 = B[['ID','Team1','innings','Team1_score','extras_run']].merge(D,left_on = 'ID',right_on='ID',how='left').drop('ID',axis=1)
ext1 = stand1[['Team1','extras_run_y']].reset_index().groupby(['Team1'])[('extras_run_y')].sum().reset_index()
ext2 = stand1[['Team2','extras_run_x']].reset_index().groupby(['Team2'])[('extras_run_x')].sum().reset_index()
ext3 = ext2.rename(columns={'Team2':'Team1','extras_run_x':'extras_run_y'}) 
extra1 = pd.concat([ext1,ext3]).groupby(['Team1'])[('extras_run_y')].sum().reset_index()
extra = extra1.rename(columns={'Team1':'Teams','extras_run_y':'Extras_Given'}).sort_values(by = 'Extras_Given',ascending = False).reset_index().drop('index',axis=1)
extra.index = extra.index + 1
extra
```

### Most Wickets By Bowler

```python
wic = ball_ipl.loc[(ball_ipl['kind'] != 'run out') & (ball_ipl['kind'] != 'retired out') & (ball_ipl['kind'] != 'retired hurt')]
W = wic.groupby(['bowler'])[('isWicketDelivery')].sum().reset_index().sort_values(by = 'isWicketDelivery',ascending = False).head(10).rename(columns={'isWicketDelivery' : 'Total Wicket','bowler' : 'Bowler'}).reset_index().drop('index',axis=1)
W.index = W.index + 1
W

```

```python
plt.figure(figsize=(10,5))
plt.bar(W['Bowler'],W['Total Wicket'],color='blue',width=0.2)
plt.xticks(rotation=90)
plt.title('Bowler Vs Total Wicekts')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.show()
```
